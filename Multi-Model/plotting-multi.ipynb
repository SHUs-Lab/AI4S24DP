```python
import pandas as pd, matplotlib.pyplot as plt

LiGO_optim_epochs = 10
```

```python
df = pd.read_pickle('Wiki103_multi_model_curves')
```

```python
df.keys()
```

```python
model_tflops_dict = {'Scratch':  1.5, 'LIGO': 1.5, "Scratch3" : 1.61, "Scratch4" : 1.73}
```

```python
df_ligo_good_103 = pd.read_pickle('LiGO_Wiki103_Good_Result')
df_ligo_poor_103 = pd.read_pickle('LiGO_Wiki103_Poor_Result')
```

```python
df_ligo_good_103.keys()
```

```python
df_ligo_poor_103.keys()
```

```python
Scratch_d_curve = {i*model_tflops_dict['Scratch']: df['Scratch'][i] for i in range(len(df['Scratch']))}
stacked_curve = {i*model_tflops_dict['Scratch']: df['Stacked Model'][i] for i in range(len(df['Stacked Model']))}
stacked_average = {i*model_tflops_dict['Scratch']: df['Stacked Model AVG'][i] for i in range(len(df['Stacked Model AVG']))}
stacked_same = {i*model_tflops_dict['Scratch']: df['Stacked Model Same'][i] for i in range(len(df['Stacked Model Same']))}

interweaving_curve = {i*model_tflops_dict['Scratch']: df['Interweaved Model'][i] for i in range(len(df['Interweaved Model']))}
interweaving_average = {i*model_tflops_dict['Scratch']: df['Interweaved Model AVG'][i] for i in range(len(df['Interweaved Model AVG']))}
interweaving_same = {i*model_tflops_dict['Scratch']: df['Interweaved Model Same'][i] for i in range(len(df['Interweaved Model Same']))}

single_stacked_curve = {i*model_tflops_dict['Scratch']: df['Single Stacked Model'][i] for i in range(len(df['Single Stacked Model']))}
single_interweaving_curve = {i*model_tflops_dict['Scratch']: df['Single Interweaved Model'][i] for i in range(len(df['Single Interweaved Model']))}
                
# ligo_d_curve = {(LiGO_optim_epochs*model_tflops_dict['LIGO'])+(i*model_tflops_dict['Scratch']): df['LiGO'][i] for i in range(len(df['LiGO']))}

ligo_good = {(LiGO_optim_epochs*model_tflops_dict['LIGO'])+(i*model_tflops_dict['Scratch']): df_ligo_good_103['LiGO_Wiki103_Good_Result'][i] for i in range(len(df_ligo_good_103['LiGO_Wiki103_Good_Result']))}
ligo_poor = {(LiGO_optim_epochs*model_tflops_dict['LIGO'])+(i*model_tflops_dict['Scratch']): df_ligo_poor_103['LiGO_Wiki103_Poor_Result'][i] for i in range(len(df_ligo_poor_103['LiGO_Wiki103_Poor_Result'])-1)}
```

```python
plt.plot(Scratch_d_curve.keys(), Scratch_d_curve.values(), label='Scratch')
plt.plot(stacked_curve.keys(), stacked_curve.values(), label='Stacked Model (Attached)')
plt.plot(stacked_average.keys(), stacked_average.values(), label='Stacked Model (Average)')
plt.plot(stacked_same.keys(), stacked_same.values(), label='Stacked Model (Same)')
plt.plot(interweaving_curve.keys(), interweaving_curve.values(), label='Interweaved Model (Attached)')
plt.plot(interweaving_average.keys(), interweaving_average.values(), label='Interweaved Model (Average)')
plt.plot(interweaving_same.keys(), interweaving_same.values(), label='Interweaved Model (Same)')
# plt.plot(single_stacked_curve.keys(), single_stacked_curve.values(), label='Single Stacked Model')
# plt.plot(single_interweaving_curve.keys(), single_interweaving_curve.values(), label='Single Interweaved Model')
plt.plot(ligo_good.keys(), ligo_good.values(), label='LiGO Good')
plt.plot(ligo_poor.keys(), ligo_poor.values(), label='LiGO Poor')
# plt.plot(ligo_d_curve.keys(), ligo_d_curve.values(), label='LiGO')
plt.grid(True)

plt.legend(loc='upper center', fontsize=7.5)
plt.ylim((0, 5))
plt.xlim((0, 110))
plt.xlabel('TFLOPS', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```
